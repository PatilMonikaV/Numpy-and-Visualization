Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

The Seaborn library in Python has many plotting functions, including scatter plots, which can help visualize relationships between two variables. Here are some other plots that Seaborn can create:        
Line plot: Shows the trend of a variable over time      
Histogram: Shows the distribution of a variable         
Box plot: Shows the distribution of a variable through quartiles, with vertical lines called whiskers indicating variability outside the upper and lower quartiles        
Violin plot: Similar to a box plot, but provides a more detailed view of the distribution of the data      
Heatmap: Shows the correlation between different variables      
Bar plot: Compares a set of categorical data      
Pair plot: Creates a grid of axis such that each numeric variable in data will create a plot between each other      
Distribution plot: Creates the distribution of any continuous  data     
Linear regression plot: Draws a linear relationship as determined through regression for the continuous variables      

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.       
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset from Seaborn
fmri_data = sns.load_dataset("fmri")

# Plot a line plot for different events and regions
plt.figure(figsize=(10, 6))  # Set the figure size

# Loop through unique events
for event in fmri_data['event'].unique():
    for region in fmri_data['region'].unique():
        # Filter data for the current event and region
        subset_data = fmri_data[(fmri_data['event'] == event) & (fmri_data['region'] == region)]
        
        # Plot the line plot for timepoint vs signal
        plt.plot(subset_data['timepoint'], subset_data['signal'], label=f"{event} - {region}")

plt.xlabel('Timepoint')  # Label for the x-axis
plt.ylabel('Signal')  # Label for the y-axis
plt.title('Signal Variation over Time for Different Events and Regions')  # Title of the plot
plt.legend()  # Add legend for different event-region combinations
plt.grid(True)  # Add grid for better readability
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Plot two box plots
plt.figure(figsize=(12, 6))  # Set the figure size

# Box plot for 'pclass' vs 'age'
plt.subplot(1, 2, 1)  # Create the first subplot
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.xlabel('Passenger Class')  # Label for the x-axis
plt.ylabel('Age')  # Label for the y-axis
plt.title('Box Plot of Age by Passenger Class')  # Title of the plot

# Box plot for 'pclass' vs 'fare'
plt.subplot(1, 2, 2)  # Create the second subplot
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.xlabel('Passenger Class')  # Label for the x-axis
plt.ylabel('Fare')  # Label for the y-axis
plt.title('Box Plot of Fare by Passenger Class')  # Title of the plot

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the histogram with hue for 'cut'
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.xlabel('Price')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis
plt.title('Histogram of Price with Hue for Cut')  # Title of the plot
plt.legend(title='Cut')  # Add legend for different cuts
plt.grid(True)  # Add grid for better readability
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset from Seaborn
iris_data = sns.load_dataset("iris")

# Plot a pair plot with hue for 'species'
sns.pairplot(data=iris_data, hue='species')
plt.suptitle('Pair Plot of Iris Dataset with Hue for Species', y=1.02)  # Title of the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset from Seaborn
flights_data = sns.load_dataset("flights")

# Pivot the data to prepare for heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot a heatmap
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)
plt.xlabel('Year')  # Label for the x-axis
plt.ylabel('Month')  # Label for the y-axis
plt.title('Heatmap of Passengers by Month and Year')  # Title of the plot
plt.show()
```
