Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
titanic_df = sns.load_dataset('titanic')

# Create a scatter plot using Plotly Express
fig = px.scatter(titanic_df, x='age', y='fare', title='Scatter Plot of Age vs Fare in Titanic Dataset')
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px

# Load the tips dataset from Plotly
tips_df = px.data.tips()

# Create a histogram using Plotly Express
fig = px.histogram(tips_df, x='sex', y='total_bill', color='day', pattern_shape='smoker',
                   title='Histogram of Total Bill by Sex with Smoker Pattern and Day Color')

fig.update_layout(barmode='group')
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
import plotly.express as px

# Load the Iris dataset from Plotly
iris_df = px.data.iris()

# Create a scatter matrix plot using Plotly Express
fig = px.scatter_matrix(iris_df, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                        color='species', title='Scatter Matrix Plot of Iris Dataset with Species Color')

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

```python
import plotly.express as px
import numpy as np

# Generate some random data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a distplot using Plotly Express
fig = px.histogram(x=data, marginal="rug", title='Distplot with Rug Plot')

fig.show()
```


A distplot is a type of plot used to visualize the distribution of a univariate set of data, showing the frequency distribution of observations. It combines a histogram with a kernel density estimate (KDE) plot, providing a comprehensive view of the data's distribution.
