To create a Bokeh plot using Python code, you'll first need to install the Bokeh library if you haven't already. You can install it using pip:
Once you have Bokeh installed, you can create a Bokeh plot using Python code. Here's a simple example to create a scatter plot using Bokeh:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a new plot with a title and axis labels
p = figure(title='Simple Scatter Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a circle glyph to the plot
p.circle(x, y, size=10, color='navy', alpha=0.5)

# Display the plot
output_notebook()
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a new plot with a title and axis labels
p = figure(title='Glyphs Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circles glyph to the plot
p.circle(x, y, size=15, color='navy', alpha=0.5)

# Add squares glyph to the plot
p.square(x, y, size=15, color='red', alpha=0.7)

# Add triangles glyph to the plot
p.triangle(x, y, size=15, color='green', alpha=0.8)

# Display the plot
output_notebook()
show(p)
```


Glyphs in Bokeh are visual markers or shapes that represent data points in a plot. They can be used to create various types of plots such as scatter plots, line plots, bar plots, and more. Examples of glyphs include circles, squares, triangles, lines, and patches.

To add glyphs to a Bokeh plot, you typically use specific glyph methods provided by Bokeh's figure object. Here's an example to illustrate how to add glyphs to a Bokeh plot:       

In this example:

We import the necessary modules from Bokeh (figure, show, output_notebook for Jupyter notebooks).
We prepare the data for the glyphs (x and y lists).
We create a new Bokeh plot (figure) with a title and axis labels.
We add three types of glyphs to the plot using different glyph methods (circle, square, triangle) provided by the figure object.
Each glyph method allows us to specify properties such as size, color, and transparency (alpha) for the glyphs.
Finally, we display the plot using show(p).

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

You can customize the appearance of a Bokeh plot in various ways, including the axes, title, legend, colors, fonts, grid lines, and more. Bokeh provides a range of options and properties that you can use to customize the visual aspects of your plot. Here's a breakdown of how you can customize different components of a Bokeh plot:


1. Axes Customization:        

You can customize the appearance of the axes using properties such as axis_label, axis_label_text_font, axis_label_text_font_size, major_label_text_font, major_label_text_font_size, axis_line_color, axis_line_width, etc.

Example:

```python
p.xaxis.axis_label = 'X-axis Label'
p.yaxis.axis_label = 'Y-axis Label'
p.xaxis.axis_label_text_font = 'Arial'
p.yaxis.axis_label_text_font_size = '12pt'
p.xaxis.axis_line_color = 'blue'
```

2. Title Customization:
You can customize the title of the plot using properties such as title, title_text_font, title_text_font_size, title_text_color, etc.

Example:

```python
p.title.text = 'Custom Title'
p.title.text_font = 'Helvetica'
p.title.text_font_size = '16pt'
p.title.text_color = 'green'
```

3. Legend Customization:
You can customize the legend of the plot using properties such as legend, legend_title, legend_location, legend_label_text_font, legend_label_text_font_size, etc.

Example:

```python
p.legend.title = 'Legend Title'
p.legend.label_text_font = 'Arial'
p.legend.label_text_font_size = '10pt'
p.legend.location = 'top_right'
```
